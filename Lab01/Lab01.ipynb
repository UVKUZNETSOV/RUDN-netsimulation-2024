### Задание №2:

Описываем функцию генерации 1000 чисел по распределению Пуассона и визуализируем гистограмму

```python
def plot_poisson(lmb=2, n_points=1000):
    dst = np.random.poisson(lmb,n_points)
    plt.hist(dst,density=True,histtype='stepfilled')
    plt.title('Распределение Пуассона')

    print('mean', np.mean(dst)) # матожидание
    print('var', np.var(dst)) # дисперсия
```

```python
plot_poisson(5,50000)
```

### Задание №3:

Задаём сторону квадрата и описываем функцию для генерации случайной равномерно распределённой точки по аналогии с ранее выполненными действиями

```python
a = 10

def squarePoint(a):
  return [np.random.uniform(1, a) for _ in range(2)]

point = squarePoint(a)
print(point)
```

Визуализируем точку в рамках ранее заданных границ

```python
plt.figure(dpi=75)
plt.xlim(a)
plt.ylim(a)
plt.plot(point[0], point[1], 'o')
plt.show()
```
